## Step 1: Understanding the Dataset

**Dataset:** NSL-KDD (Network Security Laboratory - Knowledge Discovery in Databases)

### Attack Categories (5 Classes):
| Class | Description | Train % | Test % |
|-------|-------------|---------|--------|
| **Benign** | Normal network traffic | 53.2% | 43.1% |
| **DoS** | Denial of Service attacks | 36.5% | 20.6% |
| **Probe** | Scanning/reconnaissance attacks | 9.3% | 5.5% |
| **R2L** | Remote to Local attacks | 0.9% | 20.6% |
| **U2R** | User to Root privilege escalation | 0.09% | 0.2% |

### Critical Issue: Class Imbalance!
- U2R has only 56 samples in training (0.09%)
- R2L has only 575 samples (0.9%)
- **Solution:** Use `class_weight='balanced'` and evaluate with MCC

```python
# Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, matthews_corrcoef
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

```python
# Load the Dataset
train_path = '../data/NSL_boosted-2.csv'
test_path = '../data/NSL_ppTest.csv'

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

print(f"Training samples: {len(train_df):,}")
print(f"Test samples: {len(test_df):,}")
print(f"\nFeatures: {train_df.shape[1]}")
print(f"\nColumns: {list(train_df.columns[:10])}...")
```

```python
# Check the Attack Categories (5-class target)
print("Attack Category Distribution (Training):")
print("="*50)
target_col = 'atakcat'  # 5-class target column
class_dist = train_df[target_col].value_counts()
print(class_dist)

# Visualize
plt.figure(figsize=(10, 5))
colors = ['green', 'red', 'orange', 'purple', 'brown']
class_dist.plot(kind='bar', color=colors)
plt.title('Attack Category Distribution (Training Set)', fontsize=14)
plt.xlabel('Attack Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
for i, v in enumerate(class_dist.values):
    plt.text(i, v + 500, f'{v:,}\n({v/len(train_df)*100:.1f}%)', ha='center', fontsize=9)
plt.tight_layout()
plt.show()
```

---
## Step 2: Data Preprocessing

### Preprocessing Steps:
1. **One-hot encoding** for categorical features (protocol_type, service, flag)
2. **MinMax scaling** for numerical features (0-1 range)
3. **Label encoding** for target variable

```python
# Prepare Data
def prepare_data(train_df, test_df, target_col='atakcat'):
    # Separate features and target
    X_train = train_df.drop(columns=[target_col, 'attack', 'twoclass'], errors='ignore')
    y_train = train_df[target_col]
    X_test = test_df.drop(columns=[target_col, 'attack', 'twoclass'], errors='ignore')
    y_test = test_df[target_col]
    
    # Encode target
    le = LabelEncoder()
    y_train_enc = le.fit_transform(y_train)
    y_test_enc = le.transform(y_test)
    
    # One-hot encode categorical columns
    categorical_cols = X_train.select_dtypes(include=['object']).columns
    X_train = pd.get_dummies(X_train, columns=categorical_cols)
    X_test = pd.get_dummies(X_test, columns=categorical_cols)
    
    # Align columns
    X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
    
    # Scale features
    scaler = MinMaxScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    return X_train_scaled, X_test_scaled, y_train_enc, y_test_enc, le, X_train.columns

X_train, X_test, y_train, y_test, label_encoder, feature_names = prepare_data(train_df, test_df)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"\nClasses: {label_encoder.classes_}")
print(f"Number of features: {len(feature_names)}")
```

---
## Step 3: Model Categories & Baselines

### Assignment Structure:

| Member | Category | Baselines to Compare | Final Selection |
|--------|----------|---------------------|----------------|
| Member 1 | **Linear** | LDA, Logistic Regression, Ridge | Logistic Regression |
| Member 2 | **Ensemble** | Random Forest, Extra Trees, AdaBoost | Random Forest |
| Member 3 | **Non-Linear** | KNN, Decision Tree, SVM-RBF | KNN |

### Decision Process:
1. Train all 3 baselines with default parameters
2. Compare Accuracy and MCC (overall + per class)
3. Select the best baseline for optimization

---
## Step 4: Linear Classifier Analysis (Notebook 01)

### Baselines Compared:
- Linear Discriminant Analysis (LDA)
- Logistic Regression
- Ridge Classifier

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression, RidgeClassifier
import time

# Define Linear Models
linear_models = {
    'LDA': LinearDiscriminantAnalysis(),
    'Logistic Regression': LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42),
    'Ridge Classifier': RidgeClassifier(class_weight='balanced', random_state=42)
}

linear_results = {}
print("="*60)
print("LINEAR CLASSIFIER BASELINE COMPARISON")
print("="*60)

for name, model in linear_models.items():
    start = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    train_time = time.time() - start
    
    acc = accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    
    linear_results[name] = {'accuracy': acc, 'mcc': mcc, 'time': train_time, 'predictions': y_pred}
    
    print(f"\n{name}:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  MCC: {mcc:.4f}")
    print(f"  Training Time: {train_time:.2f}s")

# Best Linear Model
best_linear = max(linear_results.items(), key=lambda x: x[1]['mcc'])
print(f"\n{'='*60}")
print(f"BEST LINEAR BASELINE: {best_linear[0]} (MCC: {best_linear[1]['mcc']:.4f})")
print(f"{'='*60}")
```

### Decision: Logistic Regression Selected

**Why Logistic Regression?**
- Highest MCC (0.7384) among linear models
- Better handling of class imbalance with `class_weight='balanced'`
- Good interpretability through coefficients

---
## Step 5: Ensemble Classifier Analysis (Notebook 02)

### Baselines Compared:
- Random Forest
- Extra Trees
- AdaBoost

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier

# Define Ensemble Models
ensemble_models = {
    'Random Forest': RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1),
    'Extra Trees': ExtraTreesClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1),
    'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42)
}

ensemble_results = {}
print("="*60)
print("ENSEMBLE CLASSIFIER BASELINE COMPARISON")
print("="*60)

for name, model in ensemble_models.items():
    start = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    train_time = time.time() - start
    
    acc = accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    
    ensemble_results[name] = {'accuracy': acc, 'mcc': mcc, 'time': train_time, 'predictions': y_pred}
    
    print(f"\n{name}:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  MCC: {mcc:.4f}")
    print(f"  Training Time: {train_time:.2f}s")

# Best Ensemble Model
best_ensemble = max(ensemble_results.items(), key=lambda x: x[1]['mcc'])
print(f"\n{'='*60}")
print(f"BEST ENSEMBLE BASELINE: {best_ensemble[0]} (MCC: {best_ensemble[1]['mcc']:.4f})")
print(f"{'='*60}")
```

### Decision: Random Forest Selected

**Why Random Forest?**
- Highest MCC (~0.81) among ensemble models
- Excellent performance on DoS and Probe detection
- Feature importance for interpretability
- Robust against overfitting through bagging

## Step 6: Non-Linear Classifier Analysis (Notebook 03)

### Baselines Compared:
- K-Nearest Neighbors (KNN)
- Decision Tree
- SVM with RBF kernel

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

# Define Non-Linear Models (using smaller sample for SVM due to time)
nonlinear_models = {
    'KNN': KNeighborsClassifier(n_neighbors=5, n_jobs=-1),
    'Decision Tree': DecisionTreeClassifier(class_weight='balanced', random_state=42),
}

nonlinear_results = {}
print("="*60)
print("NON-LINEAR CLASSIFIER BASELINE COMPARISON")
print("="*60)

for name, model in nonlinear_models.items():
    start = time.time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    train_time = time.time() - start
    
    acc = accuracy_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)
    
    nonlinear_results[name] = {'accuracy': acc, 'mcc': mcc, 'time': train_time, 'predictions': y_pred}
    
    print(f"\n{name}:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  MCC: {mcc:.4f}")
    print(f"  Training Time: {train_time:.2f}s")

# Note: SVM-RBF is slow, results from notebook:
print("\nSVM-RBF (from notebook - slow to run):")
print("  Accuracy: 0.8429 (84.29%)")
print("  MCC: 0.7685")
print("  Training Time: ~113s")

print(f"\n{'='*60}")
print(f"BEST NON-LINEAR: KNN selected (fast + good MCC)")
print(f"{'='*60}")
```

### Decision: KNN Selected

**Why KNN over SVM-RBF?**
- SVM-RBF had slightly higher MCC (0.7685 vs 0.7602)
- BUT KNN is 16x faster (7s vs 113s)
- KNN easier to tune (k, distance metric)
- Better for practical deployment

## Step 7: Optimization Strategy

### Two Optimization Approaches:

**1. Hyperparameter Tuning (GridSearchCV)**
- Systematic search over parameter grid
- Cross-validation for robust selection

**2. Feature Selection**
- Remove highly correlated features (>0.9 correlation)
- Remove low variance features
- Reduces overfitting and training time

```python
# Example: KNN Hyperparameter Tuning
print("KNN Hyperparameter Search Space:")
print("="*50)

knn_params = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'p': [1, 2],  # 1=Manhattan, 2=Euclidean
    'algorithm': ['ball_tree', 'kd_tree']
}

print("")
print("| Parameter | Values | Justification | Reference |")
print("|-----------|--------|---------------|-----------|")
print("| n_neighbors | 3,5,7,9,11 | Odd values prevent ties | Hastie et al. (2009) |")
print("| weights | uniform, distance | Distance gives local influence | Cover & Hart (1967) |")
print("| p | 1, 2 | Manhattan vs Euclidean | Aggarwal et al. (2001) |")
print("| algorithm | ball_tree, kd_tree | Search efficiency | scikit-learn docs |")

print("\n" + "="*50)
print("Best Parameters Found (from notebook):")
print("  n_neighbors: 3")
print("  weights: distance")
print("  p: 1 (Manhattan distance)")
print("  algorithm: ball_tree")
```

```python
# Feature Selection Example
from sklearn.feature_selection import VarianceThreshold

print("Feature Selection Process:")
print("="*50)
print(f"\nOriginal features: {X_train.shape[1]}")

# Variance Threshold
selector = VarianceThreshold(threshold=0.01)
X_train_var = selector.fit_transform(X_train)
print(f"After variance filter: {X_train_var.shape[1]}")

# Correlation-based (simplified demo)
print(f"After correlation filter: ~30 features (from notebook)")
print(f"\nFeature reduction: 122 -> 30 (75.4% reduction)")
```

## Summary: Key Decisions Made

### 1. Dataset Decisions
- Used **multi-class** (5 categories) instead of binary classification
- Applied `class_weight='balanced'` to handle imbalance
- Used **MCC** as primary metric (better for imbalanced data)

### 2. Model Selection Decisions
| Category | Baselines Tested | Selected | Reason |
|----------|-----------------|----------|--------|
| Linear | LDA, LogReg, Ridge | **Logistic Regression** | Highest MCC (0.74) |
| Ensemble | RF, ExtraTrees, AdaBoost | **Random Forest** | Best overall + interpretable |
| Non-Linear | KNN, DT, SVM-RBF | **KNN** | Fast + good MCC |

### 3. Optimization Decisions
- **Hyperparameter tuning** via GridSearchCV with 5-fold CV
- **Feature selection** reduced features by 75% (122 -> 30)
- **Manhattan distance** (p=1) chosen for KNN (better for high-dimensional data)

### 4. Final Winner: KNN
- **Accuracy:** 87.52%
- **MCC:** 0.8161
- **Balanced** performance across all attack classes
- **Fast** prediction time for real-time deployment
