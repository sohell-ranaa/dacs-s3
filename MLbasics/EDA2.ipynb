```python
# Descriptive Statistics
tomatjus_pd.describe(include='all').transpose()
```

```python
# Display rows with missing data
tomatjus_pd[tomatjus_pd.isnull().any(axis=1)]
```

```python
# Count the number of missing values for each column
tomatjus_pd.isnull().sum()
total = tomatjus_pd.isnull().sum().sort_values(ascending=False)
percent = (tomatjus_pd.isnull().sum()/tomatjus_pd.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, 
                         keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
# Count the number of non-missing data
tomatjus_pd.notna().sum()
```

```python
# Drop missing values
# tomatjus_pd.shape
# tomatjus_pd = tomatjus_pd.dropna()
# tomatjus_pd.shape
# Fill missing values with mean value
tomatjus_pd.fillna(tomatjus_pd.mean(), inplace = True)
tomatjus_pd.head()
```

```python
# Box Plots for all columns
fig,  ax = plt.subplots(len(list(tomatjus_pd.columns)), figsize=(8,40))

for i, feature_name in enumerate(list(tomatjus_pd.columns)):
    sns.boxplot(y=tomatjus_pd[feature_name], ax=ax[i]);
    ax[i].set_xlabel(feature_name, fontsize=8);
    ax[i].set_title("Box plot {} ".format(feature_name), fontsize=8);

plt.show();
```

```python
# Heatmap
plt.figure(figsize=(12, 9));
correlation_matrix = tomatjus_pd.corr().round(2);
sns.heatmap(correlation_matrix, cmap="YlGnBu", annot=True);
```

```python
# Highlight high correlation
sns.heatmap(correlation_matrix[(correlation_matrix >= 0.5) | (correlation_matrix <= -0.4)], 
            cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1,
            annot=True, annot_kws={"size": 8}, square=True);
```

```python
# Calculate correlation
import operator

individual_features_df = []
for i in range(0, len(tomatjus_pd.columns) - 1): # -1 because the last column is quality
    tmpDf = tomatjus_pd[[tomatjus_pd.columns[i], 'quality']]
    individual_features_df.append(tmpDf)

all_correlations = {feature.columns[0]: feature.corr()['quality'][0] for feature in individual_features_df}
all_correlations = sorted(all_correlations.items(), key=operator.itemgetter(1))
for (key, value) in all_correlations:
    print("{:>15}: {:>15}".format(key, value))
```
